```python
import os
from lxml import etree

def mathml2latex_yarosh(equation):
    """ MathML to LaTeX conversion with XSLT from Vasil Yaroshevich """
    xslt_file = os.path.join('..', 'mathconverter', 'xsl_yarosh', 'mmltex.xsl')
    dom = etree.fromstring(equation)
    xslt = etree.parse(xslt_file)
    transform = etree.XSLT(xslt)
    newdom = transform(dom)
    return str(newdom)

activation_function = """<math xmlns="http://www.w3.org/1998/Math/MathML">
  <semantics>
    <mrow>
      <mi>f</mi>
      <mo stretchy="false">(</mo>
      <mi>x</mi>
      <mo stretchy="false">)</mo>
      <mo>=</mo>
      <mfrac>
        <mn>1</mn>
        <mrow>
          <mo stretchy="false">(</mo>
          <mn>1</mn>
          <mo>+</mo>
          <msup>
            <mi>e</mi>
            <mrow>
              <mi>h</mi>
              <mi>x</mi>
            </mrow>
          </msup>
          <mo stretchy="false">)</mo>
        </mrow>
      </mfrac>
    </mrow>
  </semantics>
</math>"""

error_function = """
<math xmlns="http://www.w3.org/1998/Math/MathML">
  <semantics>
    <mrow>
      <mi>l</mi>
      <mo>=</mo>
      <mfrac>
        <mn>1</mn>
        <mn>2</mn>
      </mfrac>
      <mo stretchy="false">(</mo>
      <mi>t</mi>
      <mi>a</mi>
      <mi>r</mi>
      <mi>g</mi>
      <mi>e</mi>
      <mi>t</mi>
      <mo>&#x2212;</mo>
      <mi>o</mi>
      <mi>u</mi>
      <mi>t</mi>
      <mi>p</mi>
      <mi>u</mi>
      <mi>t</mi>
      <msup>
        <mo stretchy="false">)</mo>
        <mn>2</mn>
      </msup>
    </mrow>
  </semantics>
</math>"""

i_def = """
<math xmlns="http://www.w3.org/1998/Math/MathML">
  <semantics>
    <mrow>
      <msub>
        <mi>I</mi>
        <mn>1</mn>
      </msub>
      <mo>=</mo>
      <mn>0.05</mn>
      <mo separator="true">,</mo>
      <msub>
        <mi>I</mi>
        <mn>2</mn>
      </msub>
      <mo>=</mo>
      <mn>0.1</mn>
    </mrow>
  </semantics>
</math>
"""

weights = """
<math xmlns="http://www.w3.org/1998/Math/MathML">
  <semantics>
    <mrow>
      <msub>
        <mi>w</mi>
        <mn>1</mn>
      </msub>
      <mo>=</mo>
      <mn>0.15</mn>
      <mo separator="true">,</mo>
      <msub>
        <mi>w</mi>
        <mn>2</mn>
      </msub>
      <mo>=</mo>
      <mn>0.2</mn>
      <mo separator="true">,</mo>
      <msub>
        <mi>w</mi>
        <mn>3</mn>
      </msub>
      <mo>=</mo>
      <mn>0.25</mn>
      <mo separator="true">,</mo>
      <msub>
        <mi>w</mi>
        <mn>4</mn>
      </msub>
      <mo>=</mo>
      <mn>0.30</mn>
      <mo separator="true">,</mo>
      <msub>
        <mi>w</mi>
        <mn>5</mn>
      </msub>
      <mo>=</mo>
      <mn>0.40</mn>
      <mo separator="true">,</mo>
      <msub>
        <mi>w</mi>
        <mn>6</mn>
      </msub>
      <mo>=</mo>
      <mn>0.45</mn>
    </mrow>
  </semantics>
</math>"""

biases = """
<math xmlns="http://www.w3.org/1998/Math/MathML">
  <semantics>
    <mrow>
      <msub>
        <mi>B</mi>
        <mn>1</mn>
      </msub>
      <mo>=</mo>
      <mn>0.35</mn>
      <mo separator="true">,</mo>
      <msub>
        <mi>B</mi>
        <mn>2</mn>
      </msub>
      <mo>=</mo>
      <mn>0.6</mn>
    </mrow>
  </semantics>
</math>"""

print(mathml2latex_yarosh(activation_function))
print(mathml2latex_yarosh(error_function))
print(mathml2latex_yarosh(i_def))
print(mathml2latex_yarosh(weights))
print(mathml2latex_yarosh(biases))
```
